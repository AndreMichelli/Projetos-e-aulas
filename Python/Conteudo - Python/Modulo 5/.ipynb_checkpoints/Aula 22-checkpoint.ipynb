Vamos utilizar o mesmo dataset das aulas anteriores.

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

### Plots de regressão

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

Esta função cria um gráfico de dispersão junto de um gráfico linear, onde podemos observar uma parte mais clara, onde ele aponta uma parte especifica onde há maior taxa de variabilidade.

Para criar uma melhor visualização dos dados inseridos, podemos segregar os dados para que possamos criar gráficos diferentes com apenas uma linha de código, apenas por adicionar mais alguns parâmetros. Por exempo:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', col='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', row='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', col='sex',row='day')
```

### PairGrids

```python
import matplotlib.pyplot as plt
```

Para este outro tópico, vamos utilizar outro dataset, este, esta relacionado a iris de flores, onde temos algumas medidas retiradas de várias flores e suas espécies.

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

O PairGrids, é basicamente um conjunto dos gráficos que vimos anteriormente, porém, com uma manipulação mais abrangente.

```python
g = sns.PairGrid(iris)
```

Apenas utilizando o sns.PairGrid() e lhe fornecendo o dataset para trabelho, ele criará diversos gráficos vazios, onde podemos manipula-los de tal forma:

Utilizando apenas a função "map" e passando "plt.scatter" ele criará apenas gráficos de dispersão, assim como já foi visto.

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

Podemos determinar tratamentos diferentes para os gráficos, utilizando diagonais, separando em 3 partes: 

A diagonal, que separa os gráficos, e as partes de cima e de baixo ("map_upper()" e "map_lower()" respectivamente), com relação à diagonal mencionada anteriormente.

```python
g = sns.PairGrid(iris)
g = g.map_upper(plt.scatter)
g = g.map_lower(sns.kdeplot)
g = g.map_diag(sns.kdeplot)
```
