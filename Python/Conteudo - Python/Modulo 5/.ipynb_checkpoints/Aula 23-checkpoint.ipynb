## Aprofundamento no Seaborn

Este notebook conterá:
* ### Estilização de gráficos

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
```

Com essa biblioteca, podemos usar o método "set_style" para determinar o fundo de nossos gráficos de 5 maneiras diferentes.
Sendo elas:
* darkgrid
* whitegrid
* dark
* white
* ticks

Para exemplificar isto, vamos demonstrar utilizando uma plotagem simples com o Countplot. 

```python
sns.set_style('white')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('dark')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('darkgrid')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
```

Também podemos retirar as bordas presentes nas laterais dos gráficos, utilizando o método "despine()".

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
sns.despine()
```

Basicamente tudo que fazemos no Seaborn, é feito em Matplotlib, pois esta biblioteca é escrita sobre a que vimos anteriormente, logo, podemos fazer alterações nos gráficos utilizando alguns métodos do Matplotlib, assim como o tamanho da figura.

```python
sns.set_style('whitegrid')
plt.figure(figsize=(12,8))
sns.countplot(x='sex',data=tips)
```

Outra forma que podemos manipular o tamanho do gráfico, é a forma "inline", que funciona da seguinte forma.

```python
sns.lmplot(x='total_bill',y="tip",data=tips,height=3, aspect=2)
```

Também, caso seja necessário, podemos alterar o contexto do gráfico gerado, adequando ele a um tamanho de imagem para o tamanho ideal para o contexto passado, utilizando "set_context".

```python
sns.set_context('poster')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('paper')
sns.countplot(x='sex',data=tips)
```

E caso o tamanho da letra não seja o suficiente, podemos adicionar o parâmetro "font_scale", para determinarmos o tamanho desejado.

```python
sns.set_context('poster',font_scale=1.25)
sns.countplot(x='sex',data=tips)
```
