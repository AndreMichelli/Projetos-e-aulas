## Aprofundamento na Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
x = np.linspace(0,5,11)
y = x**2
```

Agora vamos falar um pouco mais sobre a gestão da figura em si.

Dentro do método "figure", podemos passar alguns parâmetros para que possamos manipular alguns aspectos da figura, sem influenciar no conteúdo do grafico.

```python
fig = plt.figure(figsize=(8,4), dpi = 100)
```

"figsize" se refere ao tamanho da figura, e "dpi" a resolução do gráfico.

```python
fig, ax = plt.subplots(figsize=(12,4))
ax.plot(x,y)
ax.set_xlabel('eixo x')
ax.set_ylabel('eixo y')
```

Caso seja útil, podemos salvar a imagem do grafico do tipo que desejamos, utilizando o método "savefig()".

```python
fig.savefig('grafico.jpg')
```

Ele irá salvar o arquivo do tipo desejado no mesmo diretório onde está salvo seu notebook ou programa.

Podemos também, incluir legendas em nossos gráficos, para que quando tivermos diversos conteúdos diferentes, fique tudo bem organizado e separado.

Para fazer isso, vamos adicionar um parâmetro na "plotagem" do gráfico, o parâmetro "label", e para que a legenda apareça, temos que fazer uso de um método de nosso eixo,  método .legend(), que irá dispor nossa legenda onde houver menos informação no gráfico, porém, podemos manipular isto, digitando 1, ele irá se dispor no canto superior direito, já digitando 2, ele irá ficar no canto superior esquerdo, e entre outras posições.

```python
fig, ax = plt.subplots(figsize=(12,4))
ax.plot(x,y,label='dados 1')
ax.set_xlabel('eixo x')
ax.set_ylabel('eixo y')
ax.legend()
```

Outro detalhe que podemos incluir no design do gráfico é outra forma de determinar cores para os dados, uma forma mais abrangente, utilizando o parâmetro "color".

```python
fig, ax = plt.subplots(figsize=(12,4))
ax.plot(x,y,label='dados 1',color='purple')
ax.set_xlabel('eixo x')
ax.set_ylabel('eixo y')
ax.legend()
```

Para aumentar ainda mais as possibilidades de cores, podemos dispor um código RGB em hexadecimal.

```python
fig, ax = plt.subplots(figsize=(12,4))
ax.plot(x,y,label='dados 1',color='#FF8C00')
ax.set_xlabel('eixo x')
ax.set_ylabel('eixo y')
ax.legend()
```

Podemos também trabalhar com a grossura da linha, fazendo uso de outro parâmetro, o "linewidth". (também pode ser escrito simplesmente "lw")

```python
fig, ax = plt.subplots(figsize=(12,4))
ax.plot(x,y,label='dados 1',color='#FF8C00', linewidth=4)
ax.set_xlabel('eixo x')
ax.set_ylabel('eixo y')
ax.legend()
```

Agora, manipularemos a tranparência da linha, com o parâmetro "alpha".

```python
fig, ax = plt.subplots(figsize=(12,4))
ax.plot(x,y,label='dados 1',color='#FF8C00', linewidth=4, alpha=0.5)
ax.set_xlabel('eixo x')
ax.set_ylabel('eixo y')
ax.legend()
```

Podemos determinar a estilização da linha a partir do parâmetro "linestyle".

```python
fig, ax = plt.subplots(figsize=(12,4))
ax.plot(x,y,label='dados 1',color='#FF8C00', linewidth=4, alpha=0.5 , linestyle='-.')
ax.set_xlabel('eixo x')
ax.set_ylabel('eixo y')
ax.legend()
```
