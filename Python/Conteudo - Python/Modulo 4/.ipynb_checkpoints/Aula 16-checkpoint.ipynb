## Introdução ao Matplotlib

Neste notebook, vamos introduzir ao curso o Matplotlib, a biblioteca de visualização de dados mais popular do Python.

```python
import matplotlib.pyplot as plt
```

É sugerido que seja colocado o Matplotlib para ser representado na linha rodada, ou seja, inline, da seguinte forma.

```python
%matplotlib inline
```

Vamos importar também o NumPy, para começar os exemplos da representação em graficos.

```python
import numpy as np
```

Agora, criaremos alguns arrays] para que seja representado.

```python
x = np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

Há diversas formas de representar os dados nos graficos.

Por enquanto iremos mostrar a forma funcional, uma forma rapida e simples de apresentação, porém sem muitas ferramentas na manipulação do grafico.

```python
plt.plot(x,y)
```

Podemos acessar alguns parâmetros de diversas funções da biblioteca para complemetar o grafico e deixa-lo mais intuitivo.

```python
plt.plot(x,y, 'r')
plt.xlabel('eixo x')
plt.ylabel('eixo y')
plt.title('titulo')
```

Agora iremos trabalhar com subplot, ou seja, subgráficos.

Vamos utilizar o método subplot(), e iremos passar 3 parâmetros para ele, primeiro, a quantidade de linhas, segundo, a quantidade de colunas e em terceiro com qual gráfico estamos trabalhando.

```python
plt.subplots(1,2)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

Agora vamos passar a ver os gráficos de forma mais avançada, como é mais comumente visto na analise de dados do python, onde poderemos manipular mais gráficos dentro de um só.

Passaremos a interpretar os gráficos como uma figura, uma caixa vazia, onde podemos preenche-la com os dados que quisermos.

```python
fig = plt.figure()
```

Na linha acima, nós basicamente criamos uma figura vazia onde podemos adicionar as informações.

Agora para adicionar informações e até gráficos embutidos dentro de outro gráfico, temos de usar os métodos que já usamos junto com o método .add_axes(), que irá adicionar eixos na figura vazia que criamos.

O método que vamos usar, o add_axes(), possui 4 parâmetros, que serão a distância de afastamento dentro do gráfico no qual está dentro, sendo o primeiro, a distancia para a esquerda, em segundo, para baixo, terceiro, para direita e o quarto, para cima. (Vale lembrar que as medidas são dadas de 0 até 1, utilizando notação decimal para representar porcentagem, e que devem ser passados dentro de uma lista)

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.3,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_xlabel('eixo x')
axes1.set_ylabel('eixo y')
axes2.set_xlabel('eixo x')
axes2.set_ylabel('eixo y')
```

Outra maneira de nós criarmos gráficos com esta biblioteca, é utilizando a forma orientada a objetos.

Fazendo uso da função subplots(), podemos criar diversos eixos, por exemplo:

```python
fig, ax = plt.subplots()
```

Na linha acima nós criamos um gráfico com apenas um eixo, porém sem conteudo algum.

```python
fig, ax = plt.subplots()
ax.plot(x,x**3,'r-.')
ax.plot(x,x**4,'b--')
```

Neste caso, os parâmetros passados para o plot são, primeiro, o eixo no qual será representado, segundo, o conteúdo do gráfico a ser representado, e por último, a estilização da linha do gráfico.

o objeto "ax" na verdade é uma lista, que mostra sua utilidade quando utilizarmos sub-gráficos, da seguinte maneira.

Também, para criar sub-gráficos utilizando esta metodologia, vamos precisar adicionar parâmetros à função subplots. Sendo elas "nrows" e "ncols", uo seja, número de linhas e colunas.

```python
fig, ax = plt.subplots(nrows=1,ncols=2)
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2)
ax[0].plot(x,x**3,'r-.')
ax[1].plot(x,x**4,'b--')
```

Uma dica que pode vir a ser útil em um ponto onde tenham muitos sub-gráficos a ponto de eles se sobreporem um sobre o outro, é o uso da função tight_layout(), que organizará o layout dos gráficos.

```python
fig, ax = plt.subplots(nrows=5,ncols=5)
```

```python
fig, ax = plt.subplots(nrows=5,ncols=5)
plt.tight_layout()
```
